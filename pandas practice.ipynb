```python
Sales=pd.read_csv(path+'Sales.tsv', sep='\t',encoding='utf-8',header=None)
```

```python
Sales.head()
```

```python
Sales.columns=['Invoice_ID','Customer_ID','Item_Summary','Batch_ID']
```

```python
def unpack(df, column):
    df[column]=df[column].apply(lambda x : dict(eval(x)))
    new_df=df[column].apply(pd.Series)
    result = pd.concat([df,new_df], axis=1).drop(column, axis=1)
    return result
```

```python
New_Sales=unpack(Sales,'Item_Summary')
```

```python
New_Sales.head()
```

```python
New_Sales.columns
```

```python
Prod_logs=pd.read_csv(path+'Production_logs.tsv', sep='\t',encoding='utf-8',header=None)
```

```python
Prod_logs.head()
```

```python
Prod_logs.columns=['Prod_Unit_ID','Batch_ID','Items_Produced','Items_Discarded']
```

```python
Prod_logs['Items_Produced']=Prod_logs['Items_Produced'].apply(lambda x : dict(eval(x)))
```

```python
Prod_items_produced=Prod_logs['Items_Produced'].apply(pd.Series)
```

```python
len(Prod_logs)
```

```python
Prod_items_produced.columns=['Items_produced_Paper','Items_produced_Rock','Items_produced_Scissor']
```

```python
Prod_logs['Items_Discarded']=Prod_logs['Items_Discarded'].apply(lambda x : dict(eval(x)))
Prod_items_discarded=Prod_logs['Items_Discarded'].apply(pd.Series)
```

```python
Prod_items_discarded.head()
```

```python
Prod_items_discarded.columns=['Items_discarded_Paper','Items_discarded_Rock','Items_discarded_Scissor']
```

```python
New_Prod_Logs=pd.concat([Prod_logs,Prod_items_produced,Prod_items_discarded], axis=1).drop(['Items_Produced','Items_Discarded'], axis=1)
```

```python
 New_Prod_Logs.head()
```

```python
Complaints=pd.read_csv(path+'Complaints.tsv', sep='\t',encoding='utf-8',header=None)
```

```python
Complaints.head()
```

```python
len(Complaints)
```

```python
Complaints.columns=['Invoice_Id','Defective_Item']
```

```python
Sales_Complaints=pd.merge(New_Sales,Complaints,how='inner',left_on='Invoice_ID',right_on='Invoice_Id',left_index=False, right_index=False)
```

```python
Sales_Complaints.head()
```

```python
Prod_Sales_Complaints=pd.merge(New_Prod_Logs,Sales_Complaints,how='inner',on='Batch_ID')
```

```python
Prod_Sales_Complaints=Prod_Sales_Complaints.fillna(0)
```

```python
New_df=Prod_Sales_Complaints.groupby(['Prod_Unit_ID'])['paper','rock','scissor'].apply(lambda x : x.astype(int).sum())
```

```python
New_df.head()
```

```python
New_df=New_df.add_suffix('_Sum').reset_index()
```

```python
New_df_prod=New_Prod_Logs.groupby(['Prod_Unit_ID'])['Items_produced_Paper','Items_produced_Rock',
                                                            'Items_produced_Scissor','Items_discarded_Paper',
                                                            'Items_discarded_Rock','Items_discarded_Scissor'].apply(lambda x : x.astype(int).sum())
```

```python
New_df_prod=New_df_prod.add_suffix('_Sum').reset_index()
```

```python
Report1=pd.merge(New_df_prod,New_df,on='Prod_Unit_ID',how='inner')
```

```python
Report1.head()
```

```python
Report1['rock']=((Report1.rock_Sum+Report1.Items_discarded_Rock_Sum)/Report1.Items_produced_Rock_Sum)*100
```

```python
Report1['paper']=((Report1.paper_Sum+Report1.Items_discarded_Paper_Sum)/Report1.Items_produced_Paper_Sum)*100
```

```python
Report1['scissor']=((Report1.scissor_Sum+Report1.Items_discarded_Scissor_Sum)/Report1.Items_produced_Scissor_Sum)*100
```

```python
Report1.head()
```

```python
Report1_final=Report1[['Prod_Unit_ID','rock','paper','scissor']]
```

```python
Report1_final=pd.melt(Report1_final,id_vars=["Prod_Unit_ID"], 
                 var_name="Item", value_name="Defect %")
```
